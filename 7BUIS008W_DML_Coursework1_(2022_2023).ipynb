```python

#create a dataframe with all training data except the target column
X = df.drop(columns=['Intervention Urgency'])
#separate target values
Y = df['Intervention Urgency'].values
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
df_scaled_X = scaler.fit_transform(X.to_numpy())
df_scaled_X = pd.DataFrame(df_scaled_X, columns=['AGE_years', 'BMI', 'HEIGHT_cm', 'WEIGHT_kg','WAIST_CIRCUMFERENCE_cm','DIASTOLIC_BLOOD_PRESSURE_mmHg',
'SYSTOLIC_BLOOD_PRESSURE_mmHg','WALK_DURATION_PER_DAY_MINUTES','COMPUTER_USE_TIME_PER_DAY_HOURS','SMOKING_STATUS','CIGARETTES_CONSUMED_PER_DAY'])
print("Scaled Dataset Using MinMaxScaler")
df_scaled_X.head()
```

**KNN CLASSIFIER**

```python
# KNN classifier
from sklearn.model_selection import train_test_split
#split dataset into train and test data
X_trainKNN, X_testKNN, Y_trainKNN, Y_testKNN = train_test_split(df_scaled_X, Y, test_size=0.2, random_state=1, stratify=Y)

from sklearn.neighbors import KNeighborsClassifier
# Create KNN classifier
knn = KNeighborsClassifier(n_neighbors = 30)
# Fit the data to classifier 
knn.fit(X_trainKNN,Y_trainKNN)
#prediction
Y_predKNNNORMAL=knn.predict(X_testKNN)
#check accuracy of our model on the test data
knn.score(X_testKNN, Y_testKNN)
```

**Gaussian NB**

```python
#gaussian NB
from sklearn.naive_bayes import GaussianNB
nb = GaussianNB()
#split dataset into train and test data
X_trainNB, X_testNB, Y_trainNB, Y_testNB = train_test_split(df_scaled_X, Y, test_size=0.2, random_state=1, stratify=Y)
nb.fit(X_trainNB, Y_trainNB)
Y_predNB = nb.predict(X_testNB)
```

```python
#GaussianNB score
print("Naive Bayes score: ",nb.score(X_testNB, Y_testNB))
```

**Accuracy NB**

```python
from sklearn.metrics import accuracy_score
print('Accuracy: {:.2f}'.format(accuracy_score(Y_testNB, Y_predNB)))
```

**Decision Tree**

```python
# decision tree
# Load libraries
import pandas as pd
from sklearn.tree import DecisionTreeClassifier # Import Decision Tree Classifier
from sklearn.model_selection import train_test_split # Import train_test_split function
from sklearn import metrics #Import scikit-learn metrics module for accuracy calculation
```

```python
# Create Decision Tree classifer object
clf = DecisionTreeClassifier()
# Train Decision Tree Classifer
X_trainDT, X_testDT, Y_trainDT, Y_testDT = train_test_split(df_scaled_X, Y, test_size=0.2, random_state=1, stratify=Y)
clf = clf.fit(X_trainDT,Y_trainDT)

#Predict the response for test dataset
Y_predDT = clf.predict(X_testDT)
```

**Accuracy DT**

```python
# Model Accuracy, how often is the classifier correct?
print("Accuracy:",metrics.accuracy_score(Y_testDT, Y_predDT))
```

**ANN**

```python

from sklearn.neural_network import MLPClassifier
# Import necessary modules
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from math import sqrt
from sklearn.metrics import r2_score
```

```python
print(df.shape)
df.describe().transpose()
```

```python
X_trainANN, X_testANN, Y_trainANN, Y_testANN = train_test_split(df_scaled_X, Y, test_size=0.2, random_state=87, stratify=Y)
```

```python
mlp = MLPClassifier(hidden_layer_sizes=(11,11,11), activation='relu', solver='adam', max_iter=500, random_state=900)
mlp.fit(X_trainANN,Y_trainANN)
Y_predANN = mlp.predict(X_testANN)
```

```python
# Model Accuracy, how often is the classifier correct?
print("Accuracy:",metrics.accuracy_score(Y_testANN, Y_predANN))
```

**GridSearchCV on ANN**

```python
%%time

from sklearn.model_selection import GridSearchCV

params = {'activation': ['relu', 'tanh', 'logistic', 'identity','exponential','sigmoid'],
          'hidden_layer_sizes': [(13,), (50,100,), (50,75,100,)],
          'solver': ['adam', 'sgd', 'lbfgs'],
          'learning_rate' : ['constant', 'adaptive', 'invscaling'],
          'max_iter': [500]
         }

mlp_classif_grid = GridSearchCV(MLPClassifier(random_state=200), param_grid=params, n_jobs=-1, cv=5, verbose=5)
mlp_classif_grid.fit(X_trainANN,Y_trainANN)

print('Train Accuracy : %.3f'%mlp_classif_grid.best_estimator_.score(X_trainANN, Y_trainANN))
print('Test Accuracy : %.3f'%mlp_classif_grid.best_estimator_.score(X_testANN, Y_testANN))
print('Best Accuracy Through Grid Search : %.3f'%mlp_classif_grid.best_score_)
print('Best Parameters : ',mlp_classif_grid.best_params_)
```

```python
mlp1 = MLPClassifier(activation= 'identity', hidden_layer_sizes= (50,75,100), learning_rate='adaptive', solver='sgd', max_iter=500, random_state=800)
mlp1.fit(X_trainANN,Y_trainANN)
```

```python
predictions = mlp1.predict(X_testANN)
```

```python
# Model Accuracy, how often is the classifier correct?
print("Accuracy:",metrics.accuracy_score(Y_testANN,predictions ))
```
